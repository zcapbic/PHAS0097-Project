## 1.1 Extracting thermal conductivity from logfile
Log for thermal conductivity:

    print        "average conductivity: $k[W/mK] @ $T K, ${ndens} /A\^3"
    average conductivity: 0.316415655473469[W/mK] @ 70 K, 0.0257443666020476 /A\^3
    
so split the string such that we extract the number out only. Then the code checks to see if it contains that string (the thermal conductivity), and if it does, it filters it out. There's also a guard in case it doesn't find the thermal conductivity, which I guess is nice and can stop any potential errors.

```python
def extract_TC(file_path):
    """Extracts the thermal conductivity (TC) value from the logfile,
    assuming that the LAMMPS script has already been run.
    
    Inputs:
    file_path - path to the logfile we want to read
    
    Outputs:
    thermal_conductivity - thermal conductivity extracted from the logfile (if it is there)"""
    
    # Open file in Python to read all lines
    with open(file_path, 'r') as file:
        content = file.read()

    # Looks for lines of the form {average conductivity: X[W/mk] @ 70 K, Y /A\^3}
    pattern = re.compile(r'average conductivity:\s*([\d.]+)\s*\[W/mK\]', re.IGNORECASE)
    # Search logfile for that line
    TC_lines = pattern.findall(content)
    
    # If the line exists, extract thermal conductivity. Inform user if otherwise
    if len(TC_lines)>0:
        thermal_conductivity = float(TC_lines[0])
        return thermal_conductivity
    else:
        return 'Thermal conductivity not found for this file'
```

## 1.2 Updating the seed for the random velocities in LAMMPS
This must be the 'consideration of many seeds' that we were talking about. I'm assuming that taking many seeds for random generation would make the experiment more accurate. I like the idea.

In the LAMMPS input script (and actually the logfile too), we have the line

    velocity     all create $T 102486 mom yes rot yes dist gaussian

which initialises the velocities for the problem. The line of interest is the first line, which randomly generates the velocities from a normal distribution. The random generation seed is `102486` and we want to change this seed.

The one caveat is that this code assumes the velocity initialisation line is of the form

    velocity     all create $T $r mom yes rot yes dist gaussian

where `r` is a pre-defined seed, so `variable r equal {...}`. So I should be careful about that, but it's not much of a problem, all I would need to do is add a variable definition line.

Once again for my use case there is no reason to change this function, as my velocity initialisations should be the same.

```python
def update_seed(file_path, new_seed):
    """Updates the seed for the velocity initialisation command in the LAMMPS input script.
    The velocity in LAMMPS is randomly initialised using a given seed.
    
    Inputs:
    file_path - path to the LAMMPS script which will be updated
    new_seed - new seed to be used in velocity initialisation
    
    Outputs:
    Updated file, if amendments have been made"""
    
    # Looks for lines of the form {variable r equal X}
    pattern = re.compile(r'^\s*variable\s+r\s*equal\s*[\d.]+\s*$')
    
    # Open the file in read mode to read all lines
    with open(file_path, 'r') as file:
        lines = file.readlines()

    # Open the file in write mode to update the lines iteratively
    with open(file_path, 'w') as file:
        for line in lines:
            # If the pattern exists, update the line with the new seed in place
            if pattern.match(line):
                line = f"variable r equal {new_seed}\n"
            
            # Write the line to the file
            file.write(line)
```

## 1.4 Updating dimensonality of LAMMPS input script
One other thing which the LAMMPS input script has inside it is the dimensonality of the problem. This can be defined in a line like follows

    dimensonality X
    
where `X` would be either `2` or `3`. I anticipate that this project will be considering two-dimensional structures quite frequently. So, we will need to be able to change the dimensions in the LAMMPS input script very easily. The function below does exactly that, using the same pattern recognition techniques as above.

```python
def update_dimensionality(file_path, new_dimension):
    """Updates the dimensionality written in the LAMMPS input script.
    
    Inputs:
    file_path - path to the LAMMPS script which will be updated
    
    Outputs:
    Updated file, if any amendments have been made"""
    
    # Look for lines of the form {dimension X}
    pattern = re.compile(r'^\s*dimension\s+(\d+)\b(?=\s*(?:#|$))')
    
    # Open the file in read mode to read all lines
    with open(file_path, 'r') as file:
        lines = file.readlines()

    # Open the file in write mode to update the lines iteratively
    with open(file_path, 'w') as file:
        # Update any line which matches line pattern
        for line in lines:
            if pattern.match(line):
                line = f"dimension {new_dimension}\n"
            
            file.write(line)
```

# 2 Creating and plotting layered lattices
Now I will define the three functions used to create and plot the lattices. The first function generates the individual layer, the second function stacks the layers, and the third function visualises it in three dimensions. I could have also used VMD but I don't have it installed yet.

## 2.1 Generating individual layers
LAMMPS has a weird way of working in two dimensions. It insists that the z coordinates must be zero, not the x/y. Therefore, if I wish to do stacking in 2D, I can't stack it over z as then the z coordinate would never be zero. This means that I have to stack in either x or y, and I chose to stack in y. In other words, the layers will be defined within the x-z plane.

```python
def generate_layer(num_cells, spacings):
    """Creates the individual layers for a lattice, in two dimensions.
    Assumes a simple cubic structure for now, with one atom per unit cell.
    
    Inputs:
    num_cells - number of cells as a two-dimensional NumPy vector [n_i,n_j]
    spacings - lattice vector components in i and j directions, 2D Numpy Vector [di,dj]
    
    Outputs:
    positions - positions i,j of the atoms within the layer"""
    
    # Create grid of positions
    x,z = np.meshgrid(range(num_cells[0]),range(num_cells[1]))
    # One atom per unit cell, spread across two dimensions
    num_atoms = num_cells**2
    # Extract spacings and create positions
    dx,dz = spacings
    positions=np.stack((x*dx,z*dz),axis=-1).reshape(-1,2)
    
    return positions
```

```python
def stacking_layers(num_layers, num_cells_per_layer, spacings, num_types):
    """Stacks layers created by the generate_layer function together, in the third direction.

    Inputs:
    num_layers - number of layers we want to stack
    num_cells_per_layer - number of cells per layer, as a two-dimensional NumPy vector [n_i, n_j]
    spacings - lattice vector components in the i,j,k directions, 3D NumPy vector [di,dj,dk]

    Outputs:
    positions - positions of the layered structure of atoms
    atom_types - types of each atom based on the layer pattern
    """

    # ----- Defining positions -----
    # Store positions
    positions = []
    # Extract spacings
    dx, dy, dz = spacings

    # ----- Defining types -----
    pattern = np.linspace(1,num_types,num_types)
    layer_types = np.tile(pattern, num_layers)
    layer_types = layer_types[:num_layers]
    # Store types
    atom_types = []
    
    # Loop over the amount of layers we want to add
    for layer_index in range(num_layers):
        layer_positions_2d = generate_layer(num_cells_per_layer, (dx, dz))
        # Add third direction to the 2D positions and shift by layer index
        layer_positions_3d = np.hstack((layer_positions_2d[:, 0].reshape(-1, 1), np.full((layer_positions_2d.shape[0], 1), layer_index * dy), layer_positions_2d[:, 1].reshape(-1, 1)))
        positions.append(layer_positions_3d)

        # Assign the type of the current layer to all atoms in this layer
        current_layer_type = layer_types[layer_index]
        atom_types.extend([current_layer_type] * layer_positions_2d.shape[0])

    # Stack the positions up
    positions = np.vstack(positions)
    atom_types = np.array(atom_types)
    atom_types = np.int64(atom_types)

    return positions, atom_types
```

```python
def lattice_plot(spacings, num_cells_per_layer, num_layers, num_types):
    # Create 3D figure
    fig = plt.figure(figsize=(8,8))
    ax = fig.add_subplot(111, projection='3d')
    
    # Extract positions
    positions, atom_types = stacking_layers(num_layers, num_cells_per_layer, spacings, num_types)

    # Scatter plot for lattice points
    scatter = ax.scatter(positions[:, 0], positions[:, 1], positions[:, 2], c=atom_types)

    # Also plot planes corresponding to the layers, this will require meshgrid
    x = np.linspace(np.min(positions[:,0]), np.max(positions[:,0]),10)
    z = np.linspace(np.min(positions[:,2]), np.max(positions[:,2]),10)
    X,Z = np.meshgrid(x,z)
                      
    # Much like stacking_layers function, go up the layers iteratively
    for layer in range(num_layers):
        Y = layer*spacings[1]*np.ones_like(X) # could equivalently use Y
        ax.plot_surface(X, Y, Z, alpha=0.07,color='red')

    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')
    ax.set_zlabel('$z$ stacking direction')
    ax.set_title('Lattice plot \n Purple: Type 1, Yellow: Type 2')
    ax.view_init(elev=10, azim=-15)
    plt.show()
```

## 3.1 Creating a LAMMPS-compatible datafile
If we want to import a datafile into the LAMMPS input script, we would have to use a command of the form

    read_data       data.txt

where `data.txt` is our datafile which contains the properties and spatial distributions of the atoms. The *eim* example in LAMMPS is one which uses this datafile reading method to solve its problem. This datafile is structured as:

    [N] atoms
    [N] atom types
    
    [Xmin] [Xmax] xlo xhi
    [Ymin] [Ymax] ylo yhi
    [Zmin] [Zmax] zlo zhi
    
    Masses
    1 [Mass1]
    2 [Mass2]
    
    Atoms
    1 1 [X1] [Y1] [Z1]
    2 2 [X2] [Y2] [Z2]
    ...
    N 1 [XN] [YN] [ZN]
    
    Velocities
    1   [Vx1] [Vy1] [Vz1]
    2   [Vx2] [Vy2] [Vz2]
    ...
    N   [VxN] [VyN] [VzN]
    
where `N`, `Xmin`, etc. are all variables which we can define when creating this datafile. Exactly the same procedure follows for our use case. The mass definitions aren't needed here, they can be defined within the LAMMPS input script itself.

I actually think the normal ordering type in the datafile is something like

    ID, group, type, charge, x, y, z
    
but of course in our case `group` and `charge` are both zero. So this would be good to keep in mind.

```python
def layered_datafile_creation(spacings, num_cells_per_layer, num_layers, num_types):
    """Creates a datafile of an arbitrary crystallographic material which is LAMMPS compatible.
    Uses the stacking_layers functions to stack the layers up.
    See the markdown above where this function is defined for details on the datafile structure.
    Alternativiely, view the eim example in LAMMPS.
    
    Inputs:
    num_layers - number of layers we want to stack
    num_cells_per_layer - number of cells per layer, as a two-dimensional NumPy vector [n_i, n_j]
    spacings - lattice vector components in the i,j,k directions, 3D NumPy vector [di,dj,dk]
    num_types - number of types of atoms
    
    Output:
    datafile.txt - datafile in LAMMPS format defining the structure"""

    # Find positions, types, IDs and group into data
    positions, atom_types = stacking_layers(num_layers, num_cells_per_layer, spacings, num_types)
    num_atoms = len(atom_types)
    IDs = np.array(range(1, num_atoms+1))
    data = np.column_stack((IDs, atom_types, positions))

    # Will take the form [ID] [Type] [X1] [Y1] [Z1] column-repeated many times
    np.savetxt('datafile.txt', data, fmt = '%i %i %s %s %s',header = 
f'''# Title

{len(data)} atoms

{len(np.unique(data[:,1]))} atom types

0.0 {max(positions[:,0])} xlo xhi
0.0 {max(positions[:,1])} ylo yhi
0.0 {max(positions[:,2])} zlo zhi

Atoms
''', comments = '')
    
    return
```
